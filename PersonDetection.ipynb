## Create dataframe from json

```python
import json
import os
import cv2
import torch
import pandas as pd
```

```python
with open("./labels.json") as fcc:
    data = json.load(fcc)
```

```python
dframe_dict = {i["id"]: {"file_name": i["file_name"], "bbox": []}
               for i in data["images"]}

for i in data["annotations"]:
    if i["category_id"] == 1:
        if i['image_id'] in dframe_dict.keys():
            dframe_dict[i["image_id"]]["bbox"].append(i["bbox"])
```

```python
dframe = pd.DataFrame([[i, dframe_dict[i]["file_name"], dframe_dict[i]["bbox"]]
                       for i in dframe_dict.keys()], columns=["id", "file_name", "bbox"]).sort_values(by="id").reset_index(drop=True)
```

```python
dframe["path"] = dframe["file_name"].apply(lambda x: os.path.join("data", x))
dframe = dframe.drop(["file_name"], axis=1)
```

## Normalize Coco annotations

center_x, center_y, width, height

```python
def normalize_(x, file_path):
    h, w, _ = cv2.imread(file_path).shape
    return [[(s[0]+(s[2]/2))/w,
             (s[1]+(s[3]/3))/h,
             s[2]/w,
             s[3]/h] for s in x]

dframe["bbox"] = dframe.apply(lambda x: normalize_(x["bbox"], x["path"]), axis=1)
```

## Test-Train Split

```python
test = dframe.sample(frac=0.1, axis=0)
train = dframe.drop(index=test.index)
test = test.reset_index(drop=True)
train = train.reset_index(drop=True)
```

## Print to Labels

```python
!rm -rf ./training/; mkdir -p "./training/labels" && cp -r "data/" "./training/images"
```

```python
def printLabels(df):
    for i in df.index:
        filename = f'{df["id"].iloc[i]:012}.txt'
        with open(os.path.join("./training","labels",filename), "w") as outfile:
            for arr in df["bbox"].iloc[i]:
                outfile.write(f"0 {arr[0]} {arr[1]} {arr[2]} {arr[3]}\n")
            outfile.close()

printLabels(train)
```
